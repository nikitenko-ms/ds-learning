# Преобразование данных

**1.** Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def eduction_transformation(education):
    """Данная функция предназначена для преобразования
    информации об образовании соискателя.

    Args:
        education (string/object): Уровень образования, год выпуска, ВУЗ, специальность

    Returns:
        string: строка со значением уровня образования
    """
    education = education.split()
    
    if 'Высшее' in education[:2] and 'образование' in education[:2]:
        return 'высшее'
    elif 'Неоконченное' in education[:2] and 'высшее' in education[:2]:
        return 'неоконченное высшее'
    elif 'Среднее' in education[:2] and 'специальное' in education[:2]:
        return 'среднее специальное'
    elif 'Среднее' in education[:2] and 'образование' in education[:2]:
        return 'среднее'

# с помощью функции eduction_transformation() формируем новый признак уровня образования в исходном DF
hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(eduction_transformation)

# переводим данные из признака "Образование" в тип "category"
hh_data['Образование'] = hh_data['Образование'].astype('category')
# выводим статистику по количеству значений признака "Образование"
display(hh_data['Образование'].value_counts())

# удаляем признак 'Образование и ВУЗ'
hh_data = hh_data.drop('Образование и ВУЗ', axis=1)
```

**3.** Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def work_experience(work):
    """Данная функция предназначена для преобразования
    информации об опыте работы соискателя.

    Args:
        work (string/object): Опыт работы: n лет m месяцев, периоды работы в различных компаниях

    Returns:
        int: строка со значением опыта работы, выраженным в месяцах
    """
    if work is np.nan:
        return np.nan
    elif work == 'Не указано':
        return np.nan
    else:
        work = work.split()
        if work[3] in ['лет', 'год', 'года'] and work[5] in ['месяц', 'месяца', 'месяцев']:
            return int(work[2]) * 12 + int(work[4])
        elif work[3] in ['лет', 'год', 'года'] and work[5] not in ['месяц', 'месяца', 'месяцев']:
            return int(work[2]) * 12
        else:
            return int(work[2])

# с помощью функции work_experience() формируем новый признак "Опыт работы (месяц)"
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(work_experience)

# удаляем признак 'Опыт работы'
hh_data = hh_data.drop('Опыт работы', axis=1)

# выведем первые пять значений нового признака
display(hh_data['Опыт работы (месяц)'].head())
# выведем медианный опыт работы
display(hh_data['Опыт работы (месяц)'].median())
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород',
                  'Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 
                  'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']

def city(city):
    """Данная функция предназначена для преобразования
    информации о городе проживания соискателя.

    Args:
        city (string/object): Город, (метро), готовность к переезду (города для переезда), 
        готовность к командировкам

    Returns:
        string: строка со значением города
    """
    city = city.split()
    
    if city[0] == 'Москва':
        return 'Москва'
    elif city[0] == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif city[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другой'
    
# с помощью функции city() формируем новый признак "Город"
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(city)

# перерводим данные из признака "Пол" в тип "category"
hh_data['Город'] = hh_data['Город'].astype('category')
# выведем распределение соискателей по городам
display(round((hh_data['Город'].value_counts(normalize=True) * 100), 2))
```

```python
def relocation(location):
    """Данная функция предназначена для преобразования
    информации о городе проживания соискателя.

    Args:
        location (string/object): Город, (метро), готовность к переезду (города для переезда), 
        готовность к командировкам

    Returns:
        string: строка со значением города
    """
    location = location.replace('(', ',')
    location = location.replace(')', ',')
    location = location.split(',')
           
    if ' готов к переезду ' in location or ' готова к переезду ' in location or ' хочу переехать ' in location:
        return True
    else:
        return False
        
# с помощью функции relocation() формируем новый признак "Готовность к переезду"
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(relocation)

# выведем распределение соискателей по готовности к переезду
display(round((hh_data['Готовность к переезду'].value_counts(normalize=True) * 100), 2))      
```

```python
def business_trip(trip):
    """Данная функция предназначена для преобразования
    информации о городе проживания соискателя.

    Args:
        trip (string/object): Город, (метро), готовность к переезду (города для переезда), 
        готовность к командировкам

    Returns:
        string: строка со значением города
    """
    trip = trip.replace('(', ', ')
    trip = trip.replace(')', ',')
    trip = trip.split(', ')
           
    if 'не готов к командировкам' in trip or 'не готова к командировкам' in trip:
        return False
    else:
        return True
        
# с помощью функции relocation() формируем новый признак "Готовность к командировкам"
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(business_trip)

# выведем распределение соискателей по готовности к командирвкам
display(round((hh_data['Готовность к командировкам'].value_counts(normalize=True) * 100), 2))

# удаляем признак 'Город, переезд, командировки'
hh_data = hh_data.drop('Город, переезд, командировки', axis=1)
```

**5.** Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость

* полный день, удаленная работа

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:

Нам бы хотелось видеть заработную плату в единой валюте, например, в рублях. Возникает вопрос, а где взять курс валют по отношению к рублю?

На самом деле язык Python имеет в арсенале огромное количество возможностей получения данной информации, от обращения к API Центробанка, до использования специальных библиотек, например pycbrf. Однако, это не тема нашего проекта.

Например, MDF.RU, данный ресурс позволяет удобно экспортировать данные о курсах различных валют и акций за указанные периоды в виде csv файлов. Мы уже сделали выгрузку курсов валют, которые встречаются в наших данных за период с 29.12.2017 по 05.12.2019. Скачать ее вы можете **на платформе**

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:
* "currency" - наименование валюты в ISO кодировке,
* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
exchange_data = pd.read_csv('data/ExchangeRates.csv')
# переводим даты в таблицах в формат DateTime
hh_data['Обновление резюме'] = (pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True)).dt.date
exchange_data['date'] = (pd.to_datetime(exchange_data['date'], dayfirst=True)).dt.date
# удалим лишние столбцы из exchange_data
exchange_data = exchange_data.drop(columns=['per','time', 'vol'], axis=1)

# с помощью lambda-функции формируем новый временный признак "Сумма ЗП"
hh_data['Сумма ЗП'] = hh_data['ЗП'].apply(lambda x: x.split()[0])
# с помощью lambda-функции формируем новый признак "Валюта ЗП"
hh_data['Валюта ЗП'] = hh_data['ЗП'].apply(lambda x: x.split()[1])

def currency_to_iso(currency):
    """Функция приводит обозначение валюты в соответствии
    со стандартом ISO.

    Args:
        currency (string/object): обозначение валюты
     Returns:
        string: обозначение валюты в соответстии со стандартом ISO
    """
    if currency == 'руб.':
        return 'RUB'
    elif currency == 'бел.руб.':
        return 'BYN'
    elif currency == 'грн.':
        return 'UAH'
    elif currency == 'сум':
        return 'UZS'
    else:
        return currency

# с помощью функции currency_to_iso() преобразуем обозначение валют
hh_data['Валюта ЗП'] = hh_data['Валюта ЗП'].apply(currency_to_iso)

# для объединения таблиц, переименовываем столбцы в exchange_data
exchange_data.rename(columns={'currency': 'Валюта ЗП', 'date': 'Обновление резюме'}, inplace=True)
# для объединения таблиц, перерводим данные в тип "object"
hh_data['Обновление резюме'] = hh_data['Обновление резюме'].astype('object')
exchange_data['Обновление резюме'] = exchange_data['Обновление резюме'].astype('object')

# объединяме таблицы hh_data и exchange_data
merged = hh_data.merge(exchange_data, 
                       on=['Обновление резюме', 'Валюта ЗП'],
                       how='left')
# для расчета ЗП в рублях, перерводим данные в тип "float64"
merged['ЗП (руб)'] = merged['Сумма ЗП'].astype('float64')
# заменяем на константу пропуски курса и пропорции для данных в рублях 
merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)
# "переворачиваем" пропорцию для умнодения
merged['proportion'] = merged['proportion'].apply(lambda x: 1/x)
# переводим в рубли все записи о ЗП
merged['ЗП (руб)'] = merged['close'] * merged['proportion'] * merged['ЗП (руб)']

# переносим признак в исходный DataFrame
hh_data['ЗП (руб)'] = merged['ЗП (руб)']

# удаляем признак "ЗП"
hh_data = hh_data.drop('ЗП', axis=1)
# удаляем признак "ЗП"
hh_data = hh_data.drop('Сумма ЗП', axis=1)
# удаляем признак "ЗП"
hh_data = hh_data.drop('Валюта ЗП', axis=1)

# выводим медианную зарпату
display(hh_data['ЗП (руб)'].median())
```

Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

```python
# зададим функцию поиска выбросов при помощи метода Z-отклонений
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    """Функция предназначена для выявления "выьросов" с помощью 
    метода Z-отклонений.

    Args:
        data (DataFrame): Исходные данные для анализа
        feature (String): Наимаенование признака для анализа
        log_scale (bool, optional): Логарифмический масштаб
        left (int, optional): Левая граница
        right (int, optional): Правая граница

    Returns:
        outlers (DataFrame): Количество выбросов
        cleaned (DataFrame): Количество оставшихся значений
    """
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
        
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    
    return outliers, cleaned

# используем функцию на исходных данных и выведем результат
outliers, cleaned = outliers_z_score_mod(hh_data, 'Возраст', log_scale=True, right=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')

# выведем таблицу с выбросами
display(outliers)
```
